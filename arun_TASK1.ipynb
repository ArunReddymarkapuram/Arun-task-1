Model will go thought the data and will recognize the patterns that are in the data. According to the patterns it will recomend the student to study a perticular chapter that he is week in. Model knows the students weakest chapter through the pattern of mistakes he made in given questions. Model does this using regression algorithm 

```python
#import data 
data=pd.read_csv('arun_TASK1.txt', names=['q1','q2','q3','q4','q5'])
```

Above code is to read the datasets from the csv file and name the columns

```python
#map data into arrays
on=np.asarray([1,0,0])
tw=np.asarray([0,1,0])
tr=np.asarray([0,0,1])
data['q5'] = data['q5'].map({'chapone': on, 'chaptwo': tw,'chaptri':tr})
```

Above code is to name the specific example, in our case names of chapters and giving them numerical identities 

```python
data
```

```python
#shuffle the data
data=data.iloc[np.random.permutation(len(data))]
```

```python
data
```

```python
# resets the indexes of the shuffeled data
data=data.reset_index(drop=True)
```

```python
data
```

```python
# splitting the data
#training data
x_input=data.ix[0:10,['q1','q2','q3','q4']]
temp=data['q5']
y_input=temp[0:11]
#test data
x_test=data.ix[11:14,['q1','q2','q3','q4']]
y_test=temp[11:15]
```

```python
#placeholders and variables. input has 4 features and output has 3 classes
x=tf.placeholder(tf.float32,shape=[None,4])
y_=tf.placeholder(tf.float32,shape=[None, 3])
#weight and bias
W=tf.Variable(tf.zeros([4,3]))
b=tf.Variable(tf.zeros([3]))
```

Placeholder simply allocates block of memory for future use. Later, we can use feed_dict to feed the data into placeholder. By default, placeholder has an unconstrained shape, which allows you to feed tensors of different shapes in a session. 
Variable in tensorflow as a normal variables which we use in programming languages. We initialize variables, we can modify it later as well. 

```python
# model 
#softmax function for multiclass classification
y = tf.nn.softmax(tf.matmul(x, W) + b)
```

Softmax function is normaly used in the last layer of a neural network, normaly network with a cross entropy in it to finalize the output.As our model is regression we use softmax to choose the final output.

```python
#loss function
cross_entropy = tf.reduce_mean(-tf.reduce_sum(y_ * tf.log(y), reduction_indices=[1]))
```

cross entropy is propability distribution, we calculate the propability of occurence of a possibility in our case the weakest chaper of a perticular student.

```python
#optimiser -
train_step = tf.train.AdamOptimizer(0.01).minimize(cross_entropy)
#calculating accuracy of our model 
correct_prediction = tf.equal(tf.argmax(y,1), tf.argmax(y_,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

The Adam optimization algorithm is an extension to stochastic gradient descent that has recently seen broader adoption for deep learning applications in computer vision and natural language processing.

```python
#session parameters
sess = tf.InteractiveSession()
#initialising variables
init = tf.initialize_all_variables()
sess.run(init)
#number of interations
epoch=2000
```

sessions are the diffrent blocks of code interactive session makes it simple and easy to write code in diffrent blocks and run them at the convenince and makes it easy to understand the code.

```python
for step in range(epoch):
   _, c=sess.run([train_step,cross_entropy], feed_dict={x: x_input, y_:[t for t in y_input.as_matrix()]})
   if step%100==0 :
       print(c)
```

Above shows learning rate for every 100 steps
